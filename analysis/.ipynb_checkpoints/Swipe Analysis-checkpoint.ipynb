```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib notebook

df = pd.read_csv("swipes_20_real_samples_100_after_10_000_epochs.csv", header=None, delimiter=";", decimal=",")
number_of_real_samples = 20

#df = pd.read_csv("swipes_5_real_5_every_500_epochs_after_5_000.csv", header=None, delimiter=";", decimal=",")
#number_of_real_samples = 5
```

```python
df_real = df.iloc[:number_of_real_samples,:]
```

```python
df_real_long = df_real.melt()
df_real_long = df_real_long[df_real_long["variable"] < 33]
#df_real_long
```

```python
df_fake = df.iloc[number_of_real_samples:,:]
```

```python
df_fake_long = df_fake.melt()
df_fake_long = df_fake_long[df_fake_long["variable"] < 33]
df_fake_long["variable"] = df_fake_long["variable"] + 0.5
#df_fake_long[df_fake_long["variable"]==0.5]
```

```python
df_long = pd.concat([df_real_long, df_fake_long], ignore_index=True)
df_long["variable"] = df_long["variable"] + 1
df_long_first_16_features = df_long[df_long["variable"] < 17]
df_long_last_16_features = df_long[df_long["variable"] >= 17]
```

```python
f, ax = plt.subplots(1,1,figsize=(10,6))
plt.xticks(rotation=90)
ax = sns.boxplot(data=df_long_first_16_features, x="variable", y="value")
```

```python
f, ax = plt.subplots(1,1,figsize=(10,6))
plt.xticks(rotation=90)
ax = sns.boxplot(data=df_long_last_16_features, x="variable", y="value")
```
