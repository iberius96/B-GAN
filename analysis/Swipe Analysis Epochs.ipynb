```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#%matplotlib notebook

#df = pd.read_csv("swipes_20_real_samples_100_after_10_000_epochs.csv", header=None, delimiter=";", decimal=",")
#number_of_real_samples = 20

df = pd.read_csv("swipes_5_real_5_every_500_epochs_after_5_000.csv", header=None, delimiter=";", decimal=",")
number_of_real_samples = 5
```

```python
df[34] = ""
for row in range(len(df.index)):
    if row < 5:
        df.iloc[row,34] = "real"
    elif row < 10:
        df.iloc[row,34] = "Epoch_5000"
    elif row < 15:
        df.iloc[row,34] = "Epoch_5500"
    elif row < 20:
        df.iloc[row,34] = "Epoch_6000"
    elif row < 25:
        df.iloc[row,34] = "Epoch_6500"
    elif row < 30:
        df.iloc[row,34] = "Epoch_7000"
    elif row < 35:
        df.iloc[row,34] = "Epoch_7500"
    elif row < 40:
        df.iloc[row,34] = "Epoch_8000"
    elif row < 45:
        df.iloc[row,34] = "Epoch_8500"
    elif row < 50:
        df.iloc[row,34] = "Epoch_9000"
    elif row < 55:
        df.iloc[row,34] = "Epoch_9500"
df
```

```python
df_long = df.melt([34])
df_long = df_long[df_long["variable"] < 33]
df_long
```

```python
df_long_list = []
for index in range(1,33):
    df_long_list.append(df_long[df_long["variable"]==index])
```

```python
f, ax = plt.subplots(1,1,figsize=(10,6))
ax = sns.boxplot(data=df_long_list[0], x=34, y="value")
plt.xticks(rotation=90)
ax.set_title(f'feature {1}')
```

```python
for index, df_long in enumerate(df_long_list):
    f, ax = plt.subplots(1,1,figsize=(10,6))
    ax = sns.boxplot(data=df_long, x=34, y="value")
    plt.xticks(rotation=90)
    ax.set_title(f'feature {index+1}')

```
